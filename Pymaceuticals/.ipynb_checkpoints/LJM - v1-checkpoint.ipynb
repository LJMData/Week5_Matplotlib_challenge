## Observations and Insights 

Observations and Insights 

Only one mouse had duplicated data and needed to be removed from the final data set, however not all mice lived to the end of the study (which can be seen by the timepoints not all totalling 45) so it may be worth revisiting this data using only mice who survied the whole of the study. 

9 treatment options were tested with a placebo which could be used for comparasion. Capomulin and Ramicane had the lowest tumour sizes across the study with the variance and SEM. It would be benifical to compare the Tumor size and timepoint 0 and the end time point to see which had the largest change in tumor size before using this information to draw conculsions. The numbers of male and female mice used were balanced across the study. 

Propriva had the lowest number of timepoints, suggesting that more of their mice did not survive the study than other groups, but further data would be required to attribute this directly to the performance of the drug.

When looking at the 4 most promising drug trials there is only one outlier which suggests that this data is reliable and reproducable 

The correlation factor between weight and tumor size is 0.95 in the mice who recieved Capomulin, which is classed as a strong correlation. This suggests that a larger body mass will lead to having larger tumors which should be considered when thinking about dosage when considering the next stage of trials. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
df=pd.merge(mouse_metadata,study_results,on="Mouse ID")

# Display the data table for preview
df
```

```python
# Checking the number of mice.
Mouse_uni = (df['Mouse ID'].unique().tolist())
Mouse_no = len(Mouse_uni)
print('The number of mice used was ' + str(Mouse_no))
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
bool_series = df.duplicated(['Mouse ID', 'Timepoint'])
mouse_dup = (bool_series[bool_series].index.values)

mouse_dup_list = mouse_dup.tolist()

mouse_new = df.iloc[mouse_dup_list]
mouse_new = mouse_new['Mouse ID'].unique()

print('The mouse with the duplicate ID is ' + str(*mouse_new))
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
df.iloc[mouse_dup_list]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_clean = df.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep='first')
mouse_clean
```

```python
# Checking the number of mice in the clean DataFrame.
Mouse_clean_uni = (mouse_clean['Mouse ID'].unique().tolist())
Mouse_clean_no = len(Mouse_uni)
print('The number of mice used was ' + str(Mouse_no))
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

#Create summary statisics serise for each requested measure
mouse_clean_mean = mouse_clean.groupby(by=["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
mouse_clean_med = mouse_clean.groupby(by=["Drug Regimen"]).median()["Tumor Volume (mm3)"]
mouse_clean_var = mouse_clean.groupby(by=["Drug Regimen"]).var()["Tumor Volume (mm3)"]
mouse_clean_std = mouse_clean.groupby(by=["Drug Regimen"]).std()["Tumor Volume (mm3)"]
mouse_clean_sem = mouse_clean.groupby(by=["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

#Assembmle into a DataFrame
Mouse_final = pd.DataFrame({"Mean" : mouse_clean_mean, "Median" : mouse_clean_med, "Variance" : mouse_clean_var, "St Dev" : mouse_clean_std, "SEM" : mouse_clean_sem})

#Format
Mouse_final["Mean"]=Mouse_final["Mean"].map("{:,.2f}".format)
Mouse_final["Median"]=Mouse_final["Median"].map("{:,.2f}".format)
Mouse_final["Variance"]=Mouse_final["Variance"].map("{:,.2f}".format)
Mouse_final["St Dev"]=Mouse_final["St Dev"].map("{:,.2f}".format)
Mouse_final["SEM"]=Mouse_final["SEM"].map("{:,.2f}".format)
Mouse_final
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line

mouse_agg = mouse_clean.groupby(by='Drug Regimen')["Tumor Volume (mm3)"]
round(mouse_agg.agg(['mean', 'median', 'var', 'std', 'sem']),2)
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

weight = mean_vol_2.iloc[:,0]
tum_vol = mean_vol_2.iloc[:,1]
correlation = st.pearsonr(weight,tum_vol)
print(f"The correlation between both factors is {round(correlation[0],2)}")


x_values=mean_vol_2["Weight (g)"]
y_values=mean_vol_2["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight of Mouse in g')
plt.ylabel('Average Vol of Tumor in mm3')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
